```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog
from fractions import Fraction
```

```python
def plot(p, q):
    fig = plt.figure(figsize=(10, 10))
    ax1 = fig.add_subplot(211)
    ax2 = fig.add_subplot(212)
    ax1.stem(range(1, len(p) + 1), p)
    ax2.stem(range(1, len(q) + 1), q)
def format_answer(cost, p, q):
    p1 = []
    q1 = []
    for i in p:
        p1.append(Fraction.from_float(float('{:.10f}'.format(i))).limit_denominator())
    for i in q:
        q1.append(Fraction.from_float(float('{:.10f}'.format(i))).limit_denominator())
    cost1 = Fraction.from_float(float('{:.10f}'.format(cost))).limit_denominator()
    return (cost1, p1, q1)
def print_answer(cost, p, q):
    print("cost: ")
    print(cost)
    print("p: ")
    print(*p)
    print("q: ")
    print(*q)
```

```python
def nash_equilibrium(a):
    c = [-1 for i in range(0, a.shape[1])]
    b = [1 for i in range(0, a.shape[0])]
    q = linprog(c, a, b).x
    p = linprog(b, -a.transpose(), c).x
    opt_sum = 0
    for i in p:
        opt_sum+=i
    cost = 1/opt_sum
    return (cost, p*cost, q*cost)
```

```python
#смешанная оптимальная стратегия
matrix = np.array([
    [4,  0, 6, 2,  2, 1],
    [3,  8, 4, 10, 4, 4],
    [1,  2, 6, 5,  0, 0],
    [6,  6, 4, 4, 10, 3],
    [10, 4, 6, 4,  0, 9],
    [10, 7, 0, 7,  9, 8]
])
cost,p,q = nash_equilibrium(matrix)
plot(p,q)
cost1,p1,q1 = format_answer(cost,p,q)
print_answer(cost1,p1,q1)
```

```python
#смешанная оптимальная стратегия
matrix = np.array([
    [2, 5, 3],
    [3, 1, 7],
    [8, 0, 2]
])
cost,p,q = nash_equilibrium(matrix)
plot(p,q)
cost1,p1,q1 = format_answer(cost,p,q)
print_answer(cost1,p1,q1)
```
