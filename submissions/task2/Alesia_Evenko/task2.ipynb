```python
# -*- coding: utf-8 -*-
from scipy.optimize import linprog
from matplotlib import pyplot
import numpy
import random
%matplotlib inline

def nash_equilibrium(a1):
    a1 = numpy.array(a1)
    n, m = a1.shape
    mina = a1.min()
    a1 = a1 - mina + 1
    a2 = -a1.transpose()
    c1 = [1] * n
    c2 = [-1] * m
    res1 = linprog(c1, a2, c2)
    res2 = linprog(c2, a1, c1)
    v = 1/(res1.fun)
    return {'v':v + mina - 1, 'p':v*res1.x, 'q':v*res2.x}

def visualization(p):
    pyplot.axis([0, len(p) + 1, 0, max(p) + 1/2])
    colors = ["green", "gold", "red", "blue", "pink", "coral", "darkred", "darkgreen"]
    random.shuffle(colors)
    for i in range(len(p)):
        pyplot.scatter(i + 1, p[i], alpha = 0.5, c=colors[i])
        pyplot.axvline(x = i + 1, ymin = 0, ymax = p[i] / (max(p) + 1/2), alpha = 0.5, c=colors[i])
    pyplot.show()

def run(a):
    res = nash_equilibrium(a)
    print ('V =', res['v'])
    print ('P =', res['p'])
    print ('Q =', res['q'])
    visualization(res['p'])
    visualization(res['q'])
```

```python
#равновесие Нэша в чистых стратегиях
run([[1, 3],
     [2, 4]])
```

```python
#спектр оптимальной стратегии неполон
run([[4, 0, 6, 2, 2, 1],
    [3, 8, 4, 10, 4, 4],
    [1, 2, 6, 5, 0, 0],
    [6, 6, 4, 4, 10, 3],
    [10, 4, 6, 4, 0, 9],
    [10, 7, 0, 7, 9, 8]])
```

```python
#спектр оптимальной стратегии полон
run([[1, 2],
     [2, 1]])
```

```python
run([[1, -3],
     [-2, 4]])
```

```python
run([[0, 0], [0, 0]])
```
