```python
import numpy as np
import sys, os
sys.path.append(os.path.join(sys.path[0],'..'))


import modetask

def nash_equilibrium(a, test = 0):
    minim = modetask.mini(a)
    result = modetask.sol(a, minim)
    if test:
        return result
    modetask.allprint(result)
```

```python
a = np.array([[4, 0, 6, 2, 2, 1],
               [3, 8, 4, 10, 4, 4],
               [1, 2, 6, 5, 0, 0],
               [6, 6, 4, 4, 10, 3],
               [10, 4, 6, 4, 0, 9],
               [10, 7, 0, 7, 9, 8]])

nash_equilibrium(a)
```

i. спектр оптимальной стратегии состоит из одной точки (т.е. существует равновесие Нэша в чистых стратегиях)

```python
a = np.array([[1, 0, 4],
               [5, 3, 8],
               [6, 0, 1]])

nash_equilibrium(a)
```

ii. спектр оптимальной стратегии неполон (т.е. некоторые чистые стратегии не используются)

```python
a = np.array([[1, 2, 2],
              [2, 1, 1],
              [4, 7, 0]])
nash_equilibrium(a)
```

iii. спектр оптимальной стратегии полон

```python
a = np.array([[1, 2],
              [2, 1]])
nash_equilibrium(a)
```
